Create a list to store the shopping items
* Use a while loop to create a menu of options for the user to add, remove, or view items from the list
* Use a for loop to iterate through the list of items and display them to the user
* Use the range() function to limit the number of items that can be added to the list
* Use the list, tuple, set, and dictionary data structures to store and manipulate the shopping items

```python
# Initialize an empty list, set, tuple, and dictionary to store shopping items.
shopping_list = []
shopping_set = set()
shopping_dict = {}
shopping_tuple = tuple()

# To set a limit for the number of items that can be added to the list
Maximum_Items = 6

# Using while loop to create Menu
while True:
    print("\nShopping List Menu:")
    print("1. Add Item")
    print("2. Remove Item")
    print("3. View Items")
    print("4. Exit")

# To Prompt the user to make a selection
    choice = input("Enter your choice (1, 2, 3, or 4): ")

# 1. Add Item
    if choice == '1':
        if len(shopping_list) < Maximum_Items:
            item = input("Enter the item you want to add: ")
            shopping_list.append(item)
            print(f" '{item}' added to your shopping cart\n")
        else:
            print(f" \nYou can only add up to {Maximum_Items} items to the list.")

    # 2. To Remove items from the List
    elif choice == '2':
        item = input(" Enter the item you want to remove: ")
        if item in shopping_list:
            shopping_list.remove(item)
            print(f" '{item}' removed from the shopping cart.")
        else:
            print(f" Item '{item}' not found in the shopping cart.")

    # 3. To view your items
    elif choice == '3':
        print("\nKindly view Your Shopping List:")
        for item in shopping_list:
            print(f"-{item} ")

    # 4. To Exit the program
    elif choice == '4':
        print("kindly proceed to purchase you items!")
        break
    else:
        print("Invalid choice, please try again.")
```
